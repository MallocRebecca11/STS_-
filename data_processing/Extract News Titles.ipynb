```python
def is_alphabet(uchar):
    if (uchar >= u'\u0041' and uchar<=u'\u005a') or (uchar >= u'\u0061' and uchar<=u'\u007a'):
        return True
    else:
        return False

def is_chinese(uchar):
    if uchar >= u'\u4e00' and uchar<=u'\u9fa5':
            return True
    else:
            return False
```

```python
def extract_info(path,fname): # e.g. fname = '1000100.txt'
    info = []
    soup = BeautifulSoup(open(path,encoding='utf-8'),features='html.parser')
    # get ID of the document
    ID = fname[:-4]
    
    # extract the date of the news
    date = soup.find('span',class_='story-time').contents
    date = date[0].replace('发布于','')
    
    # extract the title of the news
    #title = (soup.find('title')).get_text()
#     for i in soup.find('h1',class_='story-headline story-headline-large'):
#         print(i.string)
    print(soup.find('h1'))
    title = (soup.find('h1',class_='story-headline story-headline-large')).get_text
    #title = title[0]
    print(title)
    title = title.replace(' - FT中文网','')
    title = title.split('|')
    print(title)
    en_title = title[0]
    cn_title = title[1]
        
    # extract the content of the news
    en_content = ''
    cn_content = ''
    en_body = soup.find_all('div',class_='leftp')
    for i in en_body:
        if i.find('p') == None or (i.find('p')).get_text() == []:
            continue
        #print((i.find('p')).get_text())
#         if (i.find('p')) is Tag:
#             en_content = en_content + ((i.find('i')).contents)[0]
        en_content = en_content + (i.find('p')).get_text()
        #en_content.append((i.find('p')).contents)
    #en_content = ' '.join(en_content)
    #print(en_content)

    cn_body = soup.find_all('div',class_='rightp')
    for i in cn_body:
        if i.find('p') == None or (i.find('p')).get_text() == []:
            continue
        cn_content = cn_content + (i.find('p')).get_text()
    
    info.append(ID)
    info.append(date)
    info.append(en_title)
    info.append(cn_title)
    info.append(en_content)
    info.append(cn_content)
    return info
```

## Extract Chinese News from UIC web page

```python
from bs4 import BeautifulSoup
import requests

CN_URL = []
# first page
url = 'https://www.uic.edu.cn/news/top_stories.htm'

r = requests.get(url,verify=False)
r.encoding = r.apparent_encoding
html_contents = r.text
html_soup = BeautifulSoup(html_contents,'html.parser')

table = html_soup.find('div',class_='youtx')
for row in table.find_all('li'):
    url = row.find('a').get('href')
    url = 'https://www.uic.edu.cn' + url[2:]
    CN_URL.append(url)

# the remaining pages
pages = range(1,23)
urls = ['https://www.uic.edu.cn/news/top_stories/'+ str(i) +'.htm' for i in pages]

for page in urls:
    r = requests.get(page,verify=False)
    html_contents = r.text
    html_soup = BeautifulSoup(html_contents,'html.parser')

    table = html_soup.find('div',class_='youtx')
    
    
    for row in table.find_all('li'):
        url = row.find('a').get('href')
        url = 'https://www.uic.edu.cn' + url[5:]
        CN_URL.append(url)
```

```python
from bs4 import BeautifulSoup
import requests
# extract English News from UIC web page
EN_URL = []
# first page
url = 'https://www.uic.edu.cn/en/news/top_stories.htm'
r = requests.get(url,verify=False)
r = requests.get(url)
html_contents = r.text
html_soup = BeautifulSoup(html_contents,'html.parser')

table = html_soup.find('div',class_='youtx')


for row in table.find_all('li'):
    url = row.find('a').get('href')
    url = 'https://www.uic.edu.cn/en' + url[2:]
    EN_URL.append(url)

# the remaining pages
pages = range(1,23)
urls = ['https://www.uic.edu.cn/en/news/top_stories/'+ str(i) +'.htm' for i in pages]

for page in urls:
    r = requests.get(page)
    html_contents = r.text
    html_soup = BeautifulSoup(html_contents,'html.parser')

    table = html_soup.find('div',class_='youtx')
    
    
    for row in table.find_all('li'):
        url = row.find('a').get('href')
        url = 'https://www.uic.edu.cn/en' + url[5:]
        EN_URL.append(url)
```

```python
import re
for i in CN_URL:
    # obtain the ID of each News
    ID = re.findall(r"info/(.+?).htm",i)
    ID = ID[0][5:]
    # print(ID + '.txt')
    
    # ontain all html content of each News page
    r = requests.get(i,verify=False)
    r.encoding = 'utf-8'
    html_contents = r.text
    html_soup = BeautifulSoup(html_contents,'html.parser')
    # save the html content into txt file whose file name is corresponding to their ID
    fname = ID + '.txt'
    path = './UIC News/Chinese/' + fname
    fsave = open(path, mode = 'a',encoding='utf-8')
    fsave.write(str(html_soup))
    fsave.close()
```
